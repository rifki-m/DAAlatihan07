```python
def nilai_terkecil(deret_bilangan):
    nilai_terkecil = deret_bilangan[0]

    for nilai in deret_bilangan:
        if nilai < nilai_terkecil:
            nilai_terkecil = nilai

    return nilai_terkecil

a = [3, 20, 100, -35, 50]
print(a)
print('Nilai terkecil:', nilai_terkecil(a))
```

```python
def pencarian_beruntun(arr, x):
    # Deklarasi awal
    k = 0  # Dimulai dari indeks pertama
    n = len(arr)  # Panjang array

    # Algoritma pencarian
    while k < n and arr[k] != x:
        k += 1

    # Cek apakah elemen ditemukan
    if k < n and arr[k] == x:
        idx = k + 1  # Penyesuaian indeks agar sesuai dengan gambar (1-based index)
    else:
        idx = 0  # Jika tidak ditemukan

    return idx

# Contoh penggunaan
arr = [5, 7, 3, 9, 2]
x = 3
hasil = pencarian_beruntun(arr, x)

if hasil != 0:
    print(f"Elemen {x} ditemukan pada indeks {hasil}")
else:
    print(f"Elemen {x} tidak ditemukan")
```

```python
def selection_sort(arr):
    n = len(arr)

    for i in range(n - 1):
        imin = i

        for j in range(i + 1, n):
            if arr[j] < arr[imin]:
                imin = j

        temp = arr[i]
        arr[i] = arr[imin]
        arr[imin] = temp

    return arr

# Contoh penggunaan
arr = [64, 25, 12, 22, 11]
print("Array sebelum diurutkan:", arr)

sorted_arr = selection_sort(arr)
print("Array setelah diurutkan:", sorted_arr)
```

```python
def sequential_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return f"Elemen '{target}' ditemukan di indeks {i}."
    return f"Elemen '{target}' tidak ditemukan."

arr = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
target = 'g'
print(sequential_search(arr, target))
```

```python
def brute_force_string_match(text, pattern):
    n = len(text)      # Panjang string sumber
    m = len(pattern)   # Panjang substring (pattern)

    # Loop untuk mencari pattern dalam text
    for i in range(n - m + 1):  # Periksa dari awal hingga akhir string
        match = True
        for j in range(m):      # Bandingkan karakter pattern dengan substring
            if text[i + j] != pattern[j]:
                match = False   # Jika ada yang tidak cocok, berhenti dan geser
                break
        if match:
            return i  # Mengembalikan posisi kecocokan pertama
    return -1  # Jika tidak ada kecocokan

# Contoh penggunaan:
text = "123456789"
pattern = "56"
result = brute_force_string_match(text, pattern)

if result != -1:
    print(f'Pattern "{pattern}" ditemukan pada indeks {result}')
else:
    print(f'Pattern "{pattern}" tidak ditemukan')
```

```python
import itertools
import math
import time

# Fungsi untuk menghitung jarak antar dua titik
def distance(point1, point2):
    return math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# Fungsi untuk menghitung panjang rute
def route_length(cities, route):
    length = 0
    for i in range(len(route) - 1):
        length += distance(cities[route[i]], cities[route[i+1]])
    length += distance(cities[route[-1]], cities[route[0]])  # Kembali ke titik awal
    return length

# Algoritma brute force untuk TSP
def tsp_brute_force(cities):
    best_route = None
    min_length = float('inf')
    for route in itertools.permutations(range(len(cities))):
        length = route_length(cities, route)
        if length < min_length:
            min_length = length
            best_route = route
    return best_route, min_length

# Generate 10 titik acak (karena brute force sangat lambat untuk lebih dari 10 titik)
import random
random.seed(42)
cities = [(random.randint(0, 100), random.randint(0, 100)) for _ in range(10)]  # Ubah 10 titik sesuai kebutuhan

# Hitung waktu eksekusi brute force
start_time = time.time()
best_route, min_length = tsp_brute_force(cities)
end_time = time.time()

print(f"Best route (brute force): {best_route}")
print(f"Minimum length (brute force): {min_length}")
print(f"Time taken (brute force): {end_time - start_time} seconds")
```

```python
import math
import time

# Fungsi untuk menghitung jarak antar dua titik
def distance(point1, point2):
    return math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# Fungsi untuk menghitung panjang rute
def route_length(cities, route):
    length = 0
    for i in range(len(route) - 1):
        length += distance(cities[route[i]], cities[route[i+1]])
    length += distance(cities[route[-1]], cities[route[0]])  # Kembali ke titik awal
    return length

# Algoritma greedy untuk TSP
def tsp_greedy(cities):
    n = len(cities)
    unvisited = set(range(n))
    route = [0]  # Mulai dari kota pertama
    unvisited.remove(0)

    while unvisited:
        last = route[-1]
        next_city = min(unvisited, key=lambda city: distance(cities[last], cities[city]))
        route.append(next_city)
        unvisited.remove(next_city)

    return route, route_length(cities, route)

# Generate 20 titik acak
import random
random.seed(42)
cities = [(random.randint(0, 100), random.randint(0, 100)) for _ in range(20)]

# Hitung waktu eksekusi greedy
start_time = time.time()
best_route, min_length = tsp_greedy(cities)
end_time = time.time()

print(f"Best route (greedy): {best_route}")
print(f"Minimum length (greedy): {min_length}")
print(f"Time taken (greedy): {end_time - start_time} seconds")
```

* Brute force memberikan hasil yang lebih akurat dan optimal, tetapi tidak
efisien dari segi waktu terutama untuk jumlah titik yang lebih besar (misalnya 20 titik).
* Greedy jauh lebih cepat dan efisien, tetapi mengorbankan kualitas solusi. Panjang rutenya tidak optimal dan bisa jauh lebih buruk dibandingkan brute force, terutama pada dataset yang kompleks.

```python
import itertools
import math
import random
import matplotlib.pyplot as plt

# Fungsi untuk menghitung jarak antara dua kota
def distance(a, b):
    return math.sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2)

# Fungsi untuk menghitung total jarak dari sebuah tur
def distance_tour(tour):
    return sum(distance(tour[i], tour[i-1]) for i in range(len(tour)))

# Fungsi untuk menghasilkan semua kemungkinan tur
def alltours(cities):
    return itertools.permutations(cities)

# Fungsi untuk mencari tur dengan jarak terpendek
def shortest_tour(tours):
    return min(tours, key=distance_tour)

# Fungsi brute force untuk menghitung tur terpendek
def brute_force(cities):
    "Generate all possible tours of the cities and choose the shortest tour."
    return shortest_tour(alltours(cities))

# Fungsi untuk menghasilkan kota secara acak
def generate_cities(num_cities):
    return [(random.randint(0, 100), random.randint(0, 100)) for i in range(num_cities)]

# Fungsi untuk memvisualisasikan tur
def plot_tour(tour):
    x, y = zip(*tour)
    plt.plot(x + (x[0],), y + (y[0],), 'bo-')  # Menutup tur dengan menghubungkan kota terakhir ke kota pertama
    plt.plot(x[0], y[0], 'ro')  # Menandai titik awal
    plt.show()

# Fungsi untuk menjalankan solusi TSP dan menampilkan hasilnya
def tsp(algorithm, cities):
    tour = algorithm(cities)
    print(f"Tour length {distance_tour(tour)}")
    plot_tour(tour)

# Contoh penggunaan dengan brute force untuk 10 kota
cities = generate_cities(20)
tsp(brute_force, cities)
```

```python
import math
import random
import matplotlib.pyplot as plt

# Fungsi untuk menghitung jarak antara dua kota
def distance(a, b):
    return math.sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2)

# Fungsi untuk menghitung total jarak dari sebuah tur
def distance_tour(tour):
    return sum(distance(tour[i], tour[i-1]) for i in range(len(tour)))

# Fungsi untuk menghasilkan kota secara acak
def generate_cities(num_cities):
    return [(random.randint(0, 100), random.randint(0, 100)) for _ in range(num_cities)]

# Algoritma Greedy untuk TSP
def greedy_tsp(cities):
    start = cities[0]  # Mulai dari kota pertama
    tour = [start]
    cities = cities[1:]

    while cities:
        # Pilih kota terdekat dari kota saat ini
        nearest = min(cities, key=lambda city: distance(tour[-1], city))
        tour.append(nearest)
        cities.remove(nearest)

    return tour

# Fungsi untuk memvisualisasikan tur
def plot_tour(tour):
    x, y = zip(*tour)
    plt.plot(x + (x[0],), y + (y[0],), 'bo-')  # Menutup tur dengan menghubungkan kota terakhir ke kota pertama
    plt.plot(x[0], y[0], 'ro')  # Menandai titik awal
    plt.show()

# Fungsi untuk menjalankan solusi TSP dan menampilkan hasilnya
def tsp(algorithm, cities):
    tour = algorithm(cities)
    print(f"Tour length {distance_tour(tour)}")
    plot_tour(tour)

# Contoh penggunaan dengan greedy untuk 20 kota
cities = generate_cities(20)
tsp(greedy_tsp, cities)
```
